```python
import pandas as pd
df = pd.read_excel('Infosy_test_multocorr.xlsx')
```

# Filling NaN values with 0s to find the mean 

```python
df = df.fillna(0)
```

```python
df.head()
```

# Mean of each column

```python
mean = []
for i,j in df.items():
    sum = 0
    for items in j:
      sum = sum + items
    mean.append(sum/252)

mean
```

# Replace 0 with Nan now so to replace it with the mean

```python
df.replace(to_replace = 0,  value = pd.NA, inplace=True)
```

```python
df.head()
```

# Replace missing values with the mean of the column

```python
for i,j in df.items():
     df[i].replace(to_replace = pd.NA,  value = mean[df.columns.get_loc(i)], inplace=True)
```

# The function to find correlation matrix and remove using thresold the correlated columns

```python
import numpy as np
def drop_correlated(dataframe, thresold_abs):
    cor_matrix = dataframe.corr().abs()
    upper_tri = cor_matrix.where(np.triu(np.ones(cor_matrix.shape),k=1).astype(np.bool))
    
    to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > thresold_abs)]
    df_new = dataframe.drop(to_drop, axis=1)
    return df_new
```

## Used on the original dataset, can use the function for standardised and/pr normalised too

```python
df_new = drop_correlated(df, 0.4)
df_new
```
